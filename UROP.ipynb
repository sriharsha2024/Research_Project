```python
#importing required libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
```

```python
#importing dataset from drive 
path = "/content/drive/MyDrive/Marks Data - Dr. Sobin.csv"
ds = pd.read_csv(path)
ds.head(5)
```

```python
# checking whether any null values exist
ds.isnull().sum()
```

```python
# removing null rows
ds.dropna()
ds.info()
```

```python
ds1=ds.drop(0)
ds1.info()
```

```python
# removing rows containing absent 
ds2=ds1.drop(1774)
ds2.info()
```

```python
# converting the pandas dataframe into numpy arrays
X = ds2['Theory - Internal'].tolist()
Y1 = ds2['Theory - External'].tolist()
print(X)
Y=[float(item) for item in Y1]
print(Y)
```

```python
# plotting graph internal marks vs external marks
f = plt.figure()
# f.set_figwidth(15)
# f.set_figheight(15)
plt.scatter(X, Y,color='r', marker="*")
x = np.linspace(-5,40,100)
plt.title('Internal marks VS external marks')
plt.plot(x, x+20, 'g',)
plt.xlabel('X- internal marks')
plt.ylabel('Y- external marks')
plt.show()
```

```python
# Blue line is Y=X+15
f = plt.figure()
# f.set_figwidth(15)
# f.set_figheight(15)
plt.scatter(X, Y,color='r', marker="*")
x = np.linspace(-5,40,100)
plt.title('Internal marks VS external marks')
plt.plot(x, x+15, 'b')
plt.xlabel('X- internal marks')
plt.ylabel('Y- external marks')
plt.show()
```

```python
# Blue line is Y=X+10
f = plt.figure()
# f.set_figwidth(15)
# f.set_figheight(15)
plt.scatter(X, Y,color='r', marker="*")
x = np.linspace(-5,40,100)
plt.title('Internal marks VS external marks')
plt.plot(x, x+10, 'b')
plt.xlabel('X- internal marks')
plt.ylabel('Y- external marks')
plt.show()
```

```python
# importing sklearn library 
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# splitting the dataset into test and train 
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.25)
```

```python
# converting list into numpy arrays
X_train=np.array(X_train)
Y_train=np.array(Y_train)
X_test=np.array(X_test)
Y_test=np.array(Y_test)
```

```python
# creating a model and training the model
lr=LinearRegression().fit(X_train.reshape(-1,1) ,Y_train)
```

```python
# predicting the external marks 
y_pred= lr.predict(X_test.reshape(-1,1))
print(y_pred[:30])
res = {X_test[i]: y_pred[i] for i in range(len(X_test))}
```

```python
print("Coefficients: \n", lr.coef_)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(Y_test, y_pred))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(Y_test, y_pred))
```

```python
# Plot outputs
plt.scatter(X_test, Y_test, color="black")
plt.plot(X_test, y_pred, color="blue")
plt.show()
```
